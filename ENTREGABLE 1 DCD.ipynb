## Métodos Importantes de los Strings en Python

En Python, los strings son objetos que tienen una variedad de métodos integrados que facilitan su manipulación. A continuación, se presentan algunos de los métodos más importantes:

- `len()`: Es un método no exclusivo de strings, pero puede usarse para retornar el largo, ya sea de un sntring, lista, tupla o diccioanrio.
- `upper()`: Convierte el string a mayúsculas.

- `lower()`: Convierte el string a minúsculas.

- `capitalize()`: Convierte el primer carácter del string a mayúscula y el resto a minúsculas.

- `strip()`: Elimina los espacios en blanco al principio y al final del string.

- `replace(old, new)`: Reemplaza todas las ocurrencias del substring "old" con el substring "new".

- `split(sep)`: Divide el string en una lista de substrings basados en el separador "sep".

- `join(iterable)`: Une los elementos de un iterable (por ejemplo, una lista) en un string utilizando el string como separador.

- `find(sub)`: Devuelve el índice de la primera ocurrencia del substring "sub" en el string. Si no se encuentra, devuelve -1.

- `startswith(prefix)`: Devuelve True si el string comienza con el prefijo "prefix", de lo contrario, devuelve False.

- `endswith(suffix)`: Devuelve True si el string termina con el sufijo "suffix", de lo contrario, devuelve False.

````python
# 1)
class Person:
  def __init__(self,name,age,city):
    self.name=name
    self.age=age
    self.city=city

Pedro=Person("Pedro",5,"CDMX")
Juan=Person("Juan",32,"León")
Pablo=Person("Pablo",29,"Guadalajara")
print("Estan registrados:")
print(Pedro.name)
print(Juan.name)
print(Pablo.name)
a=input("Quién te cae mejor? ")

while True:
  if a==Pedro.name:
    print("Pedro vive en ",Pedro.city)
    print("Y tiene ",Pedro.age," años")
    break
  elif a==Juan.name:
    print("Juan vive en ",Juan.city)
    print("Y tiene ",Juan.age," años")
    break
  elif a==Pablo.name:
    print("Pablo vive en ",Pablo.city)
    print("Y tiene ",Pablo.age," años")
    break
  else:
    print("No lo conozco")
    a=input("Quién te cae mejor? ")
````

````python
# 2)
class Rectangle:
  def __init__(self,height,withd):
    self.height=height
    self.withd=withd
  def area(self):
    return float(self.height*self.withd)
  def perimetro(self):
    return float(2*(self.height+self.withd))
height=float(input("Dime la altura del rectángulo "))
withd=float(input("Dime el ancho del rectángulo "))
rect=Rectangle(height,withd)
print(f"El área del rectángulo es ",rect.area())
print(f"El perímetro del rectángulo es ",rect.perimetro())
````

 [\w\.-]+@[\w\.-]+

se usa para validar direcciones de correo electrónico. Cada parte de la expresión significa

[\w\.-]+: Busca uno o más caracteres que sean palabras (\w), puntos (.), o guiones (-).

@: Busca el símbolo de arroba.

[\w\.-]+: Repite la búsqueda anterior, buscando uno o más caracteres que sean palabras, puntos, o guiones.

Esta expresión regular busca algo con la estructura de algunaco@lgomco.com.


````python
# 2)
import re

texto="Ejemplo de dirección de correo electrónico: arroyoescalona.pablo@gmail.com"
patron=r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
coincidencia=re.search(patron,texto)

print("Tú dierección de correo es ",coincidencia)
````

````python
# 3)
texto="Teléfono de contacto: 123-456-7890, Teléfono móvil: (987) 654-3210"
patron = r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}'
coincidencia = re.findall(patron, texto)

print("Tus números de teléfono son ",coincidencia)
````

````python
# 4)

texto= "Este es un ejemplo de texto donde queremos buscar la palabra clave 'ejemplo'."
print("El ejemplo es:")
print(texto)
palabra_clave = input("Dime una palabra clave ")
patron = rf'\b{palabra_clave}+\b'
coincidencia = re.findall(patron, texto, re.IGNORECASE)

print("Las palabras clave son ",coincidencia)
print("Se repite ",len(coincidencia)," veces")
````

````python
!pip install requests beautifulsoup4
import requests
from bs4 import BeautifulSoup

url = 'https://guadalajarasecreta.com/que-hacer/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

print("Puedes realizar las siguientes actividades en Guadalajara:")
titles = soup.find_all('h2')
for title in titles:
    print(title.get_text())
````
